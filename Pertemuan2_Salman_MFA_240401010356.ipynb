```python
import pandas as pd
import numpy as np
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

```

```python
print(df.shape, df.columns.tolist())
print(df.isnull().sum())
```

```python
hasil = df.groupby('Pclass').agg(
total=('PassengerId','count'),
selamat=('Survived','mean'),).round(3)
print(hasil)
```

```python
survived_people = (df['Survived'] == 1).sum()
percent_survived = survived_people / df['Survived'].count() * 100
print(percent_survived)
```

```python
wk1 = df[(df['Sex']=='female') & (df['Pclass']==1)]
print(wk1['Age'].mean())
```

```python
hasil = df.groupby('Pclass').agg(
    total=('PassengerId','count'),
    selamat=('Survived','mean')
).round(3)
print(hasil)

hasil_only_mean = df.groupby('Pclass')['Survived'].mean()
print(hasil_only_mean)
```
